```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
sns.lineplot(data['total_bill'])
plt.show()
```

```python
sns.scatterplot(x = 'total_bill',y = 'tip', data=data)
plt.show()
```

```python
sns.histplot(data['total_bill'])
plt.show()
```

```python
plt.hist(data['total_bill'])
plt.show()
```

```python
sns.kdeplot(data['total_bill'])
plt.show()
```

```python
#catplot
#relpot
sns.relplot(x = 'total_bill',y='tip',data = data)
plt.show()
```

```python
sns.relplot(x = 'total_bill',y='tip', hue = 'time', data = data)
plt.show()
```

```python
sns.relplot(x = 'total_bill',y='tip', hue = 'time',style = 'smoker', data = data)
plt.show()
```

```python
sns.relplot(x = 'total_bill',y='tip', hue = 'time',style = 'smoker', data = data,kind = 'scatter',row ='time')
plt.show()
```

```python
sns.relplot(x = 'total_bill',y='tip', hue = 'time',style = 'smoker', data = data,kind = 'scatter',col = 'smoker')
plt.show()
```

```python
sns.relplot(x = 'total_bill',y='tip', hue = 'time',style = 'smoker', data = data,kind = 'scatter',col = 'day')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill',fontsize=15, color='red')
plt.ylabel('Tip',fontsize=15, color='green')
plt.show()
```

```python
sns.catplot(x ='day', y='total_bill',data=data)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill', kind = 'swarm',data=data)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=data)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='bar', data=data)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=data)
plt.show()
```
