**MATPLOTLIB**
| *Category*     | *Details*                                                                                      |
| ---------------- | ------------------------------------------------------------------------------------------------ |
| *Definition*   | A Python 2D plotting library used for creating static, interactive, and animated visualizations. |
| *Main Module*  | matplotlib.pyplot                                                                              |
| *Developed In* | Python                                                                                           |
| *Common Use*   | Data visualization in data science, machine learning, and research                               |










| *Plot Type*  | *Description*                                                              |
| -------------- | ---------------------------------------------------------------------------- |
| Line Plot      | Shows trends over time or continuous data using lines.                       |
| Bar Plot       | Displays categorical data with rectangular bars.                             |
| Histogram      | Shows distribution of a dataset by grouping into bins.                       |
| Scatter Plot   | Plots points for two variables to show relationships or clusters.            |
| Pie Chart      | Represents data as slices of a pie to show proportions.                      |
| Box Plot       | Shows data distribution with median, quartiles, and

```python
import numpy as np
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c = c, alpha=0.5)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c = c, alpha=0.5, cmap='coolwarm')
plt.colorbar()
plt.show()
```

### Bar Plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x, y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.barh(x, y) ## diplays horizontal bars
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x, y, color = 'green', width = 0.1) 
plt.show()
```

##pie

```python
x = np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
plt.pie(x, labels=labels)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
plt.pie(x, labels=labels,explode=myexplode)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
plt.pie(x, labels=labels,explode=myexplode, shadow=True)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
plt.pie(x, labels=labels,explode=myexplode,startangle=90,shadow=True)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
c = ['red','green','blue','yellow']
plt.pie(x, labels=labels,explode=myexplode,startangle=90,shadow=True,colors=c)
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
c = ['red','green','blue','yellow']
plt.pie(x, labels=labels,explode=myexplode,startangle=90,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
x = np.array([1,2,3,4])
labels  = ["apple","banana","cherry","dates"]
myexplode = [0,0.2,0,0]
c = ['red','green','blue','yellow']
plt.pie(x, labels=labels,explode=myexplode,startangle=90,shadow=True,autopct='%1.1f%%')
plt.legend(title = "my pie chart")
plt.show()
```
